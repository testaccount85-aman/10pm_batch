```python
import pandas as pd
```

```python
week1 = pd.read_csv(r"dataset\Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv(r"dataset\Restaurant - Week 2 Sales.csv")
food = pd.read_csv(r"dataset\Restaurant - Foods.csv")
customer = pd.read_csv(r"dataset\Restaurant - Customers.csv")
```

```python
sales = pd.concat([week1 , week2] , ignore_index=True)
```

```python
saleswithfood = sales.merge(food , how = "left" , on = "Food ID")
```

```python
data = saleswithfood.merge(customer , how = "left" , left_on="Customer ID" , right_on="ID")
```

```python
data.drop(columns=["ID"] , inplace=True)
```

```python
data.to_csv("aman.csv" , index=False)
```
