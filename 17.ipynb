```python
names = data["Name"].str.split("," , expand=True)
```

```python
names.columns = ["lastname" , "firstname"]
```

```python
names.rename(columns={"lastname":"Last Name" , "firstname":"First Name"} , inplace=True)
```

```python
data = pd.concat([names , data] , axis=1).drop(columns=["Name"])
```
